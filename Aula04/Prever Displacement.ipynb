```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cars.csv")
```

```python
df.drop(["car name"],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.corr().round(2).style.background_gradient(cmap='coolwarm', axis=None)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.drop(["displacement"],axis=1),
    df.displacement,
    test_size=0.33,
)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
model_n = LinearRegression(normalize=True)
model_n.fit(X_train,y_train)
model_n.score(X_test,y_test)
```

```python
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
from keras.utils.vis_utils import plot_model

from sklearn.metrics import r2_score
```

```python
model_nn = Sequential()
model_nn.add(Dense(7, input_dim=7, kernel_initializer='normal', activation='relu')) # camada oculta
model_nn.add(Dense(10, kernel_initializer='normal', activation='relu'))# entrada
#model_nn.add(Dense(3, kernel_initializer='normal', activation='relu'))
model_nn.add(Dense(1, kernel_initializer='normal')) # saída
model_nn.compile(loss='mean_squared_error', optimizer='adam')
#plot_model(model_nn, show_shapes=True, show_layer_names=True)
```

```python
model_nn.fit(X_train, y_train, epochs=150, batch_size=16)
```

```python
# avaliação
metrica = model_nn.evaluate(X_test, y_test)
print('Mean Squared Error: %.2f' % (metrica))
```
